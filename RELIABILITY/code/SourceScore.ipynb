```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../data/SourceInfo.csv")
```

```python
df.head()
```

```python
df["SourceScore"]=""

d ={"High":3,"Low":1,"Medium":2}

df['Year'] = df['Year'].fillna(2020)
```

```python
for i in df.index:
    df["SourceScore"][i]=0.2*df['Wikipedia'][i]+0.01*(2020-df['Year'][i])+0.1*df['Twitter'][i]+0.05*df['Verified'][i]+0.37*d[df['Popularity'][i]]+0.2*df['History'][i]+0.07*df['Controversy'][i]
```

```python
scores=np.reshape(np.asarray(df['SourceScore']),(-1,1))
scaler = MinMaxScaler()
scaler.fit(scores)
scores=scaler.transform(scores)
print(scores)

df['SourceScore']=scores
```

```python
df.to_csv("../data/SourceInfo.csv",index=None)
```
