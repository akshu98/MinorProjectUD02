```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../data/SourceInfo.csv")
```

```python
df
```

```python
df["SourceScore"]=""

d ={"High":3,"Low":1,"Medium":2}

df['Year'] = df['Year'].fillna(2020)
```

```python
for i in df.index:
    if(2020-df['Year'][i]>20):
        df["SourceScore"][i]=0.2*df['Wikipedia'][i]+(0.01*20)+0.1*df['Twitter'][i]+0.05*df['Verified'][i]+0.37*d[df['Popularity'][i]]+0.2*df['History'][i]+0.07*df['Controversy'][i]
    else:
        df["SourceScore"][i]=0.2*df['Wikipedia'][i]+0.01*(2020-df['Year'][i])+0.1*df['Twitter'][i]+0.05*df['Verified'][i]+0.37*d[df['Popularity'][i]]+0.2*df['History'][i]+0.07*df['Controversy'][i]
```
